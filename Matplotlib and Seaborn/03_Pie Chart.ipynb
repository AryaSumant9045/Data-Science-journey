```python
import matplotlib.pyplot as plt
labels = ["Virat", "Sachin", "Yuvraj", "Cahal"]
runs = [223, 443, 235, 225]

color = ["#f80ffc", "#0dff11", "#00bbff", "#f7ff0f"]
plt.pie(runs, labels = labels, colors = color)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=color,
    wedgeprops={'edgecolor': 'black', 'linewidth': 3, 'linestyle': '--'}
)
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=color)
plt.title("Exploded Pie Example")
plt.show()

plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=color,
    shadow=True  # adds a 3D-like shadow
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=color,
    shadow=True,
    startangle=140
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',# for showing percentage 
    startangle=140
)
plt.title("Career Run Share")
plt.show()
```
