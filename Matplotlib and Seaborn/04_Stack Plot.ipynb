```python
import matplotlib.pyplot as plt
activites = ['Studing', 'Playing', 'Sleeping', 'Watching TV']
colors = ["yellow", "green", "black", "red"]
explode= [0, 0.1, 0.2, 0]

time_spend = [3, 4, 2, 5]
plt.pie(time_spend, labels = activites,colors =colors, autopct="%1.1f%%", startangle = 90, explode= explode, shadow=True)
# plt.figure(figsize = (6,6)) #figsize=[width, height]
plt.title("School Studens routine")
plt.axis('equal') #aspect ratio equal hota haii.e. perfect circle bantaa hai
```

```python
import matplotlib.pyplot as plt
 
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['studying', 'playing', 'watching_tv', 'sleeping']
colors = ['green', 'yellow', 'orange', 'pink', 'skyblue']

plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors = colors, alpha=0)
plt.legend(loc='upper left')
plt.grid()
plt.title("Comparision of Student Routine over one week")
plt.xlabel("days")
plt.ylabel("Hours")
```
