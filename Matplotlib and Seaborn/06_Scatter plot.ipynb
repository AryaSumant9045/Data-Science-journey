```python
import matplotlib.pyplot as plt
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 40, 85, 90]

colors = [ "red" if color<50 else "green" for color in exam_scores] 
sizes = [ score*2 for score in exam_scores]

plt.scatter(study_hours, exam_scores, c= colors , s = sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Score Marks")
plt.title("Study hours vs scores")
```

```python

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 40, 85, 90]
# Add labels
for i in range(len(study_hours)):
    plt.annotate(f'Stud_{i+1}', (study_hours[i], exam_scores[i]))
    
plt.scatter(study_hours, exam_scores)
plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 65, 40, 85, 90]
# Add labels
for i in range(len(study_hours)):
    plt.annotate(f'Stud_{i+1}', (study_hours[i]+0.2, exam_scores[i]-0.4))
    
plt.scatter(study_hours, exam_scores)
plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label = "A", color = "blue")
plt.scatter(class_b_hours, class_b_scores , label = "B", color = "red")

 
plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
