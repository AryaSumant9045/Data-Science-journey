```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
#good practice
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli, label = "virat Kohli")
plt.plot(years, sehwag, label = "Swhwag")
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Comparision")
plt.legend()
plt.show()
           
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli,'r1--')   # format string= [color] [marker] [line]
plt.plot(years, sehwag , 'go:' )
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Comparision")
plt.legend(["virat Kohli","Swhwag"])
plt.show()
```

```python
plt.plot(years, kohli, color="green", linestyle="--", linewidth= 3,label = "Kohli")
plt.plot(years, sehwag, color = "red",linestyle = "-.",label = "sehwag" )
plt.plot(years, runs, color = "orange", linestyle = ":", label = "Sachin") 
plt.legend()
plt.tight_layout()
plt.grid(True)
```

```python
# Assignment of today lecture
import matplotlib.pyplot as plt
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
rohit = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, color = "red", linestyle = ":", linewidth = 4, label = "Virat Kohle")
plt.plot(years, rohit, color = "green", linestyle = "--", linewidth = 3, label = "Rohit Sharma")
plt.plot(years, sehwag, color = "blue", linestyle = "-.", linewidth = 2, label = "Sehwag ")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Player Comparision")
plt.grid(True)
plt.tight_layout()
```
